```python
# Animal: Integer 1-5 representing which animal they would like to adopt
# Danger: List of Floats of 5 elements representing the relative danger of a different animal to the original 5
# Cats/Dogs/Birds/Fish/Mammals: Integer representing the number of animals of each type the owner currently possesses

def funx(animal, danger, cats, dogs, birds, fish, mammals, other):
    
    m = [
        [0.20, 0.20, 0.25, 0.01, 0.40, float(danger[0])],
        [0.20, 0.10, 0.20, 0.01, 0.30, float(danger[1])],
        [0.25, 0.20, 0.05, 0.01, 0.05, float(danger[2])],
        [0.01, 0.01, 0.01, 0.01, 0.01, float(danger[3])],
        [0.40, 0.30, 0.05, 0.01, 0.01, float(danger[4])]
        ]
    
    uwdanger = 0
    danger = 0

    totalpets = cats + dogs + birds + fish + mammals + other

    for n in range(totalpets):

        if n == 0:
            continue
        
        # The probability that n altercations occur
        for a1 in range(n+1):
            for a2 in range(n-a1+1):
                for a3 in range(n-a1-a2+1):
                    for a4 in range(n-a1-a2-a3+1):
                        for a5 in range(n-a1-a2-a3-a4+1):

                            # Altercation vector is [a1, a2, a3, a4, a5, a6]
                            a6 = n-a1-a2-a3-a4-a5
                            
                            # Probability alterations occur
                            p = (m[animal-1][0]) ** a1
                            p = (m[animal-1][0]) ** a1 * (m[animal-1][1]) ** a2 * (m[animal-1][2]) ** a3 * (m[animal-1][3]) ** a4 * (m[animal-1][4]) ** a5  * (m[animal-1][5]) ** a6
                            
                            # Probability no other altercations do
                            q = (1 - m[animal-1][0]) ** (cats - a1) * (1 - m[animal-1][1]) ** (dogs - a2) * (1 - m[animal-1][2]) ** (birds - a3) * (1 - m[animal-1][3]) ** (fish - a4) * (1 - m[animal-1][4]) ** (mammals - a5) * (1 - m[animal-1][5]) ** (other - a5)

                            # Number of combinations possible:
                            r = comb(cats, a1) * comb(dogs, a2) * comb(birds, a3) * comb(fish, a4) * comb(mammals, a5) * comb(other, a6)

                            # Weight
                            s = n

                            danger += p * q * r * s
                            uwdanger += p * q * r

    return([danger, uwdanger])
```

```python
knowledgebase = {}
danger = [0.1, 0.1, 0.1, 0.01, 0.01]

def fight(a,b,c,d,e,f):
    wdanger = 0

    newkey = a * 10 ** 5 + b * 10 ** 4 + c * 10 ** 3 + d * 10 ** 2 + e * 10 + f
    if newkey in knowledgebase:
        return(knowledgebase[newkey])
    
    if a > 0:
        key = (a-1) * 10 ** 5 + b * 10 ** 4 + c * 10 ** 3 + d * 10 ** 2 + e * 10 + f
        if key in knowledgebase:
            wdanger = knowledgebase[key] + funx(1, danger, a-1, b, c, d, e, f)[0]
            knowledgebase[newkey] = wdanger
            return(wdanger)
        else:
            wdanger = fight(a-1,b,c,d,e,f) + funx(1, danger, a-1, b, c, d, e, f)[0]
            knowledgebase[newkey] = wdanger
            return(wdanger)

    if b > 0:
        key = a * 10 ** 5 + (b-1) * 10 ** 4 + c * 10 ** 3 + d * 10 ** 2 + e * 10 + f
        if key in knowledgebase:
            wdanger = knowledgebase[key] + funx(2, danger, a, b-1, c, d, e, f)[0]
            knowledgebase[newkey] = wdanger
            return(wdanger)
        else:
            wdanger = fight(a,b-1,c,d,e,f) + funx(2, danger, a, b-1, c, d, e, f)[0]
            knowledgebase[newkey] = wdanger
            return(wdanger)

    if c > 0:
        key = a * 10 ** 5 + b * 10 ** 4 + (c-1) * 10 ** 3 + d * 10 ** 2 + e * 10 + f
        if key in knowledgebase:
            wdanger = knowledgebase[key] + funx(3, danger, a, b, c-1, d, e, f)[0]
            knowledgebase[newkey] = wdanger
            return(wdanger)
        else:
            wdanger = fight(a,b,c-1,d,e,f) + funx(3, danger, a, b, c-1, d, e, f)[0]
            knowledgebase[newkey] = wdanger
            return(wdanger)

    if d > 0:
        key = a * 10 ** 5 + b * 10 ** 4 + c * 10 ** 3 + (d-1) * 10 ** 2 + e * 10 + f
        if key in knowledgebase:
            wdanger = knowledgebase[key] + funx(4, danger, a, b, c, d-1, e, f)[0]
            knowledgebase[newkey] = wdanger
            return(wdanger)
        else:
            wdanger = fight(a,b,c,d-1,e,f) + funx(4, danger, a, b, c, d-1, e, f)[0]
            knowledgebase[newkey] = wdanger
            return(wdanger)

    if e > 0:
        key = a * 10 ** 5 + b * 10 ** 4 + c * 10 ** 3 + d * 10 ** 2 + (e-1) * 10 + f
        if key in knowledgebase:
            wdanger = knowledgebase[key] + funx(5, danger, a, b, c, d, e-1, f)[0]
            knowledgebase[newkey] = wdanger
            return(wdanger)
        else:
            wdanger = fight(a,b,c,d,e-1,f) + funx(5, danger, a, b, c, d, e-1, f)[0]
            knowledgebase[newkey] = wdanger
            return(wdanger)
    
    wdanger = 0
    knowledgebase[newkey] = wdanger
    return(wdanger)
```

```python
dlist = []
for i in range(len(dfpetpet)):
    dlist.append(fight(dfpetpet.iloc[i][1], dfpetpet.iloc[i][2], int(dfpetpet.iloc[i][3]), int(dfpetpet.iloc[i][4]), int(dfpetpet.iloc[i][5]), int(dfpetpet.iloc[i][5])))
```
